```python
import scipy
from __future__ import division
```

```python
from scipy.optimize import brentq
from scipy import cosh,exp
import matplotlib.pyplot as plt
import pprint
```

```python
roots = []
for p in range(10,100,1):
    a = brentq(lambda x :cosh(x*(p-1.0)/(p+1.0)) - 0.5*exp(x),0,10)
    roots.append(a)
#print roots
#pprint.pprint(roots)
plt.style.use('ggplot')
plt.plot(range(10,100,1),roots)
plt.xlabel('p')
plt.ylabel('roots')
plt.title('roots vs p')
plt.show()
```
