```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special
```

# plot Bessel Function

```python
#dir(scipy.special)
x=np.linspace(-10,10,100)
y1=scipy.special.jn(1,x)
y2=scipy.special.jn(2,x)
y3=scipy.special.jn(3,x)
#scipy.special.jn?
ax1 = plt.plot(x,y1,x,y2,x,y3)
ax2 = plt.plot(x,x*0)
plt.xlim([-10,10])
plt.ylim([-1,1])
plt.legend((ax1),('bessel_1','bessel_2','bessel_3'))
plt.show()
```
